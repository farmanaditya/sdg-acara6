```python
import pandas as pd
import folium

# Membaca data dari CSV
data = pd.read_csv('D:/KULIAH/SEMESTER 4/Bandara.csv')

# Membuat peta dengan lokasi awal yang ditentukan secara otomatis
peta = folium.Map(location=[data['Latitude'].mean(), data['Longitude'].mean()], zoom_start=10)

# Iterasi melalui setiap baris data untuk menambahkan marker pada peta
for index, row in data.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Nama Bandara']).add_to(peta)

# Menampilkan peta
peta
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the first shapefile
first_shapefile = gpd.read_file('D:/SHAPEFILE/diy_kab/bantul.shp')

# Load the second shapefile
second_shapefile = gpd.read_file('D:/SHAPEFILE/diy_kab/gunungkidul.shp')

# Plot the shapefiles on top of each other 
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the first shapefile in red 
first_shapefile.plot(ax=ax, color='red', edgecolor='black')

# Plot the second shapefile in green, on top of the first shapefile 
second_shapefile.plot(ax=ax, color='green', edgecolor='black', alpha=0.5)

# Set the title for the subplot
ax.set_title('Bantul dan Gunungkidul')

# Show the plot 
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the first shapefile (administrative boundary)
shapefile1 = gpd.read_file("D:/SHAPEFILE/KOTA SURAKARTA/ADMINISTRASIDESA_AR_25K.shp")

# Read the second shapefile (health facilities)
shapefile2 = gpd.read_file("D:/SHAPEFILE/KOTA SURAKARTA/RUMAHSAKIT_PT_25K.shp")

# Define target CRS (UTM, for example)
target_crs = 'EPSG:32649'  # UTM zone 49S, adjust according to your area

# Convert both shapefiles to the target CRS
shapefile1 = shapefile1.to_crs(target_crs)
shapefile2 = shapefile2.to_crs(target_crs)

# Create a buffer around health facilities with appropriate radius
# Radius in meters, adjust as needed
buffer_distance = 500  # for example, 200 meters
shapefile2_buffer = shapefile2.buffer(buffer_distance)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(20, 10))

# Plot the first shapefile (administrative boundary)
shapefile1.plot(ax=ax, color='green', edgecolor='black')

# Plot the second shapefile (health facilities)
shapefile2.plot(ax=ax, color='yellow', edgecolor='black')

# Plot the buffer around health facilities
gpd.GeoSeries(shapefile2_buffer).plot(ax=ax, color='red', alpha=0.3)

# Set the title
ax.set_title('Kota Surakarta')

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Import the first shapefile
shapefile1 = gpd.read_file('D:/SHAPEFILE/KOTA SEMARANG/PEMERINTAHAN_PT_25K.shp')

# Import the second shapefile
shapefile2 = gpd.read_file('D:/SHAPEFILE/KOTA SEMARANG/RUMAHSAKIT_PT_25K.shp')

# Plot the first shapefile
plt.figure(figsize=(10, 5))
shapefile1.plot(color='blue', edgecolor='black')
plt.title('PEMERINTAHAN')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)

# Plot the second shapefile
plt.figure(figsize=(10, 5))
shapefile2.plot(color='red', edgecolor='black')
plt.title('RUMAHSAKIT')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)

# Show the plots
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the first shapefile (administrative boundary)
shapefile_admin = gpd.read_file("D:/SHAPEFILE/KOTA SEMARANG/PEMERINTAHAN_PT_25K.shp")

# Read the second shapefile (health facilities)
shapefile_hospital = gpd.read_file("D:/SHAPEFILE/KOTA SEMARANG/RUMAHSAKIT_PT_25K.shp")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the administrative boundary
shapefile_admin.plot(ax=ax, color='blue', edgecolor='black', label='Pemerintahan')

# Plot the health facilities
shapefile_hospital.plot(ax=ax, color='red', edgecolor='black', label='Rumah Sakit')

# Add legend
ax.legend()

# Set title and labels
ax.set_title('Pemerintahan dan Rumah Sakit di Kota Semarang')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show grid
ax.grid(True)

# Show the plot
plt.show()
```
